# Unification of Knowledge: Best Prompt OR Best ML Model

Ensemble of the best response generated by the LLM and prompts, along with the best ML model for classification: based on a logical OR operation

## 0 Imports

```python
import pandas as pd
import numpy as np
from sklearn.metrics import matthews_corrcoef, accuracy_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
y_pred_nb = pd.read_csv("../01_ML/y_pred_ML/y_pred_nb.csv", sep = ",")
```

```python
y_pred_GPT_03_profiled_simple_prompt = pd.read_csv("../02_LLM/y_pred_LLMs/GPT/y_pred_GPT_o3_profiled_simple_prompt.csv", sep =",") # okay, but not great, but better than Gemma alone
```

## 1 Best LLM OR Best ML Model

```python
# concat the predictions of the best LLM and the best ML model
df_models = pd.concat([y_pred_GPT_03_profiled_simple_prompt["y_pred"], y_pred_nb["y_pred"]], axis=1)

# apply logical OR operation on the predictions
y_pred_OR = pd.DataFrame({
    'y_pred': [0] * len(df_models),
})

for i in range(len(df_models)):
    if df_models.iloc[i, 0] == 1 or df_models.iloc[i, 1] == 1:
        y_pred_OR.iloc[i, 0] = 1
    else:
        y_pred_OR.iloc[i, 0] = 0
```

```python
y_pred_OR.to_csv("y_pred_unification/OR_unification/y_pred_best_LLM_or_best_ML.csv", sep = ",", index = False)
```
