```python
y_pred_Gemma_few_shot_prompt = pd.read_csv("y_pred_LLMs/Gemma/y_pred_gemma_few_shot_prompt.csv", sep = ",")
y_pred_Grok_few_shot_prompt = pd.read_csv("y_pred_LLMs/Grok/y_pred_Grok_few_shot_prompt.csv", sep = ",") # higher sensistivity but overall not better
y_pred_GPT_o3_cot_prompt = pd.read_csv("y_pred_LLMs/GPT/y_pred_GPT_o3_cot_prompt.csv", sep = ",") # better than alone
y_pred_Gemini_few_shot_prompt = pd.read_csv("y_pred_LLMs/Gemini/y_pred_Gemini_few_shot_prompt.csv", sep = ",") # works very well, migh MCC but low specificity
y_pred_Claude_few_shot_prompt = pd.read_csv("y_pred_LLMs/Claude/y_pred_Claude_few_shot_prompt.csv", sep = ",") # also works very well, migh MCC but low specificity
y_pred_DeepSeek_few_shot_prompt = pd.read_csv("y_pred_LLMs/DeepSeek/y_pred_deeps_few_shot_prompt.csv", sep = ",") # works well, but not as good as Gemini or Claude
y_pred_Claude_cot_prompt = pd.read_csv("y_pred_LLMs/Claude/y_pred_claude_cot_prompt.csv", sep = ",") # not so good
y_pred_GPT_4_profiled_simple_prompt = pd.read_csv("y_pred_LLMs/GPT/y_pred_GPT_profiled_simple_prompt.csv", sep = ",") # okay, but not great, but better than Gemma alone
```

## 1 Best LLM OR Best ML Model

```python
# concat the predictions of the best LLM and the best ML model
df_models = pd.concat([y_pred_Gemma_few_shot_prompt["y_pred"], y_pred_Gemini_few_shot_prompt["y_pred"]], axis = 1) # best so far
# df_models = pd.concat([y_pred_Gemma_few_shot_prompt["y_pred"], y_pred_dnn["y_pred"]], axis=1)

# apply logical OR operation on the predictions
y_pred_OR = pd.DataFrame({
    'y_pred': [0] * len(df_models),
})

for i in range(len(df_models)):
    if df_models.iloc[i, 0] == 1 or df_models.iloc[i, 1] == 1:
        y_pred_OR.iloc[i, 0] = 1
    else:
        y_pred_OR.iloc[i, 0] = 0
```

```python
y_pred_OR.to_csv("../exp/y_pred_unification/OR_unification/y_pred_best_LLM_or_best_ML.csv", sep = ",", index = False)
```
