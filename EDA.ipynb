```python
youth_drug_dat = pd.read_csv("dat/youth_drug/rems82be29_pd.txt", sep = "\t", low_memory = False)
youth_drug_dat
```

```python
# get column number of "ZIGGER"
col_num_1 = youth_drug_dat.columns.get_loc("ZIGGER")
col_num_2 = youth_drug_dat.columns.get_loc("ZUVIELDR")

# get column names of col_num_1 to col_num_2
col_names = youth_drug_dat.columns[col_num_1:col_num_2+1]

# calculate mean of all columns in col_names save as drug use
# recode 9 as NaN
youth_drug_dat[col_names] = youth_drug_dat[col_names].replace(9, np.nan)
youth_drug_dat[col_names] = youth_drug_dat[col_names].replace(8, np.nan)

# standardize the columns by deviding by max of column
youth_drug_dat[col_names] = youth_drug_dat[col_names].div(youth_drug_dat[col_names].max(axis=1), axis=0)

youth_drug_dat["drug_use"] = youth_drug_dat[col_names].mean(axis=1)

# describe column "drug_use"
youth_drug_dat["drug_use"].describe()
```

```python
# count values of drug use for that are > 0.5
youth_drug_dat[youth_drug_dat["drug_use"] > 0.2].count()
```
